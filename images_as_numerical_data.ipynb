# Images as Grids of Pixels

### Importing libraries

```python
import numpy as np
import matplotlib.image as mpimg  # for reading in images

import matplotlib.pyplot as plt
import cv2  # computer vision library

%matplotlib inline
```

### Reading and displaying the images

```python
# Read in the image
image = mpimg.imread('Lenna.png')

# Print out the image dimensions
print('Image dimensions:', image.shape)

plt.imshow(image);
```

```python
lenna_red_channel = image[:, :, 0]
plt.imshow(lenna_red_channel);
```

```python
lenna_green_channel = image[:, :, 1]
plt.imshow(lenna_green_channel);
```

```python
lenna_blue_channel = image[:, :, 2]
plt.imshow(lenna_blue_channel);
```

```python
# Read in the image
image = mpimg.imread('car.jpg')

# Print out the image dimensions
print('Original Image dimensions:', image.shape)

plt.imshow(image);
```

```python
# Change from color to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
print('Grayscale Image dimensions:', gray_image.shape)

plt.imshow(gray_image, cmap='gray');
```

```python
# Prints specific grayscale pixel values
# What is the pixel value at x = 400 and y = 300 (on the body of the car)?

x = 400
y = 300

print(gray_image[y,x])
```

```python
# Finds the maximum and minimum grayscale values in this image

max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print('Max: ', max_val)
print('Min: ', min_val)
```

```python
# Creating a 5x5 image using just grayscale, numerical values
tiny_image = np.array([[0, 20, 30, 150, 120],
                      [200, 200, 250, 70, 3],
                      [50, 180, 85, 40, 90],
                      [240, 100, 50, 255, 10],
                      [30, 0, 75, 190, 220]])

# To show the pixel grid, use matshow
plt.imshow(tiny_image, cmap='gray')
```

```python
# Read in the image
image = mpimg.imread('highway.jpg')

plt.imshow(image);
```
