### Gradient descent algorithms

There are three different variants of gradient descent algorithms - 
<ol>
    <li>Batch gradient descent (when we pass all the data at one go through the network)</li>
    <li>Mini batch gradient descent (when we break the data into mini batches and pass through the network one at a time)</li>
    <li>Stochastic gradient descent (when we pass the one data point through the network at a time)</li>
</ol>

### Challenges of gradient descent

1. Choosing a proper learning rate can be difficult. A learning rate that is too small leads to painfully slow convergence, while a learning rate that is too large can hinder convergence and cause the loss function to fluctuate around the minimum or even to diverge.

![alt text](gd4.png "Title")

2. The same learning rate applies to all parameter updates. If our data is sparse and our features have very different frequencies, we might not want to update all of them to the same extent, but perform a larger update for rarely occurring features.

![alt text](gd5.png "Title")

3. The problem of **saddle points**, the saddle points are usually surrounded by a plateau of the same error, which makes it hard for SGD to escape, as the gradient is close to zero in all dimensions.

![alt text](gd6.jpeg "Title")

### Gradient Descent Optimization techniques

#### 1. SGD with Momentum

SGD has trouble navigating the areas where surface curves much more steep in one dimension than in another, which are common around local optima. In these scenarios, SGD oscillates across the slopes of that area while only making hesitant progress along the bottom towards the local optimum.

Momentum helps to **get over the local minima** and also helps to **converge faster**

![alt text](gd8.png "Title")

![alt text](gd7.png "Title")

![alt text](gd9.png "Title")

#### 2. RMSProp

- Maintain a moving (discounted) average of the square of gradients
- Divides the learning rate by an exponentially decaying average of squared gradients

![alt text](gd12.png "Title")

![alt text](gd11.png "Title")

#### 3. Adam

Adaptive Moment Estimation (Adam) is another method that computes adaptive learning rates for each parameter. In addition to storing an exponentially decaying average of past squared gradients like RMSprop, Adam also keeps an exponentially decaying average of past gradients similar to momentum
